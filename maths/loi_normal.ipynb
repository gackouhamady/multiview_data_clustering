# Comprendre la Normalisation dans la Loi Normale : Pourquoi $$\sigma \sqrt{2\pi}$$ ?

## 1. Le Rôle de la Constante de Normalisation

La loi normale utilise le terme $$\frac{1}{\sigma \sqrt{2\pi}}$$ pour garantir que :
- L'aire totale sous la courbe = 1 (100% de probabilité)
- La fonction reste une vraie densité de probabilité

## 2. Origine Mathématique

### Cas Standard ($$\mu=0$$, $$\sigma=1$$)
L'intégrale fondamentale :

```math
\int_{-\infty}^{+\infty} e^{-x^2/2} \text{d}x = \sqrt{2\pi}
```

### Cas Général ($$\mu$$ et $$\sigma$$ quelconques)
Après changement de variable :

```math
\int_{-\infty}^{+\infty} e^{-(x-\mu)^2/(2\sigma^2)} \text{d}x = \sigma \sqrt{2\pi}
```

## 3. Démonstration Clé
On calcule le carré de l'intégrale :

```math
I^2 = \left(\int e^{-x^2/2} \text{d}x\right)^2 = \int\int e^{-(x^2+y^2)/2} \text{d}x \text{d}y
```

Passage en coordonnées polaires :

```math
I^2 = \int_0^{2\pi} \int_0^{\infty} e^{-r^2/2} r \text{d}r \text{d}\theta = 2\pi
```

Donc :

```math
I = \sqrt{2\pi}
```

## 4. Interprétation
| Terme  | Rôle                                |
|--------|-------------------------------------|
| $$\sigma$$  | Ajuste la largeur de la courbe      |
| $$\sqrt{2\pi}$$  | Constante provenant de l'intégrale gaussienne |
| $$\frac{1}{\sigma \sqrt{2\pi}}$$  | Normalise l'aire totale à 1         |

Cette normalisation est essentielle pour que la loi normale soit une distribution de probabilité valide.

## 5. Exemple Visuel
Courbe normale avec :

$$\mu = 0$$

$$\sigma = 1$$

Aire totale :

```math
\int_{-\infty}^{+\infty} \frac{1}{\sqrt{2\pi}} e^{-x^2/2} \text{d}x = 1
```

Conclusion : Le terme $$\sigma \sqrt{2\pi}$$ assure la cohérence probabiliste de la loi normale.
