# Comprendre la Normalisation dans la Loi Normale : Pourquoi $$\sigma \sqrt{2\pi}$$ ?

## 1. Le Rôle de la Constante de Normalisation

La loi normale utilise le terme $$\frac{1}{\sigma \sqrt{2\pi}}$$ pour garantir que :
- L'aire totale sous la courbe = 1 (100% de probabilité)
- La fonction reste une vraie densité de probabilité

## 2. Origine Mathématique

### Cas Standard ($$\mu=0$$, $$\sigma=1$$)
L'intégrale fondamentale :

$$
\int_{-\infty}^{+\infty} e^{-x^2/2} \, dx = \sqrt{2\pi}
$$

### Cas Général ($$\mu$$ et $$\sigma$$ quelconques)
Après changement de variable :

$$
\int_{-\infty}^{+\infty} e^{-(x-\mu)^2/(2\sigma^2)} \, dx = \sigma \sqrt{2\pi}
$$

## 3. Démonstration Clé
On calcule le carré de l'intégrale :

$$
I^2 = \left(\int_{-\infty}^{\infty} e^{-x^2/2} \, dx\right)^2 = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} e^{-(x^2+y^2)/2} \, dx \, dy
$$

Passage en coordonnées polaires :

$$
I^2 = \int_0^{2\pi} \int_0^{\infty} e^{-r^2/2} r \, dr \, d\theta = 2\pi
$$

Donc :

$$
I = \sqrt{2\pi}
$$

## 4. Interprétation
| Terme  | Rôle                                |
|--------|-------------------------------------|
| $$\sigma$$  | Ajuste la largeur de la courbe      |
| $$\sqrt{2\pi}$$  | Constante provenant de l'intégrale gaussienne |
| $$\frac{1}{\sigma \sqrt{2\pi}}$$  | Normalise l'aire totale à 1         |

Cette normalisation est essentielle pour que la loi normale soit une distribution de probabilité valide.

## 5. Exemple Visuel
Courbe normale avec :

$$\mu = 0$$

$$\sigma = 1$$

Aire totale :

$$
\int_{-\infty}^{+\infty} \frac{1}{\sqrt{2\pi}} e^{-x^2/2} \, dx = 1
$$

Conclusion : Le terme $$\sigma \sqrt{2\pi}$$ assure la cohérence probabiliste de la loi normale.

# Gaussian Integral Proof

We aim to prove the well-known Gaussian integral:

$$
\int_{-\infty}^{\infty} e^{-x^2/2} \, dx = \sqrt{2\pi}
$$

## Step 1: Squaring the Integral

We first define the integral:

$$
I = \int_{-\infty}^{\infty} e^{-x^2/2} \, dx
$$

To compute \( I \), we consider the square of the integral:

$$
I^2 = \left( \int_{-\infty}^{\infty} e^{-x^2/2} \, dx \right) \times \left( \int_{-\infty}^{\infty} e^{-y^2/2} \, dy \right)
$$

Since both integrals are over the entire real line and are identical, we write:

$$
I^2 = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} e^{-(x^2 + y^2)/2} \, dx \, dy
$$

## Step 2: Conversion to Polar Coordinates

We introduce polar coordinates:

$$
x = r \cos \theta, \quad y = r \sin \theta
$$

The Jacobian for the transformation is:

$$
dx \, dy = r \, dr \, d\theta
$$

Rewriting the integral in polar form:

$$
I^2 = \int_0^{2\pi} \int_0^\infty e^{-r^2/2} \, r \, dr \, d\theta
$$

## Step 3: Evaluating the Radial Integral

We evaluate the inner integral:

$$
J = \int_0^\infty e^{-r^2/2} \, r \, dr
$$

Using the substitution \( u = r^2/2 \) which implies \( du = r \, dr \), we rewrite the integral as:

$$
J = \int_0^\infty e^{-u} \, du
$$

This is a standard Gamma function integral:

$$
\int_0^\infty e^{-u} \, du = 1
$$

Thus, we obtain:

$$
J = 1
$$

Now, integrating over \( \theta \):

$$
I^2 = \int_0^{2\pi} 1 \, d\theta = 2\pi
$$

## Step 4: Taking the Square Root

Since we found $$I^2 = 2\pi$$ taking the square root gives:

$
I = \sqrt{2\pi}
$

Thus, we have proven the Gaussian integral:

$$
\int_{-\infty}^{\infty} e^{-x^2/2} \, dx = \sqrt{2\pi}
$$
